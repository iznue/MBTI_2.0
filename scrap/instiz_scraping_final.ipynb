### 인스티즈 사이트 댓글 크롤링 작업

```python
from bs4 import BeautifulSoup
from selenium import webdriver
from time import *

driver = webdriver.Chrome()
driver.get('https://www.instiz.net/pt/7276279')

comments = []

for a in range(16):
    html = driver.page_source
    # print(html)
    soup = BeautifulSoup(html, 'html.parser')

    reviews = soup.find_all('div', {'class':'comment_line'})
    
    for review in reviews:
      extract = soup.find('span', attrs={'class':'minitext'}).extract()
      print(review.get_text().strip())
      print('----------------')
      comments.append(review.get_text().strip())
    
    sleep(3)
```

```python
comments
```

```python
###########################################################################
# 문단별로 나누었을 때 첫 문장(mbti 기제)이 영어인 경우만 추출하여 데이터로 사용하기
# 사람들이 적은 mbti에 s n을 구분하여 column으로 지정
# 질문 1, 2, 3에 대한 답변을 따로 구분하여 column으로 지정

import re

# mbti를 한국어, 영어로 기제한 경우를 따로 분류함
user_comment_eng = []
user_comment_kor = []

for comment in comments:
    comment_list = comment.split('\n')

    if re.match("^[a-zA-Z]+$", comment_list[0]):
        # print(comment_list)
        # print('------------')
        user_comment_eng.append(comment_list)
    
    elif re.match("^[가-힣]+$", comment_list[0]):
        # print(comment_list)
        # print('------------')
        user_comment_kor.append(comment_list)
```

##### mbti를 영어로 기재한 경우

```python
# S, N 구분 테스트
n_cnt, s_cnt = 0,0
for com in user_comment_eng:
    com[0] = com[0].upper()

    if 'N' in com[0]:
        n_cnt += 1
    elif 'S' in com[0]:
        s_cnt += 1
print(n_cnt, s_cnt)
```

```python
mbti_comment_eng = []

for us_com_eng in user_comment_eng:
    us_com_eng[0] = us_com_eng[0].upper()

    if len(us_com_eng) == 4:
        # print(us_com)
        if 'N' in us_com_eng[0]:
            us_com_eng[0] = 'N'
            mbti_comment_eng.append(us_com_eng)
        elif 'S' in us_com_eng[0]:
            us_com_eng[0] = 'S'
            mbti_comment_eng.append(us_com_eng)
```

```python
mbti_comment_eng
```

##### mbti를 한글로 기재한 경우

```python
# S, N 구분 테스트
n_cnt_kor, s_cnt_kor = 0,0
for com_kor in user_comment_kor:
    if com_kor[0].startswith('잇') or com_kor[0].startswith('엣'):
        s_cnt_kor += 1
    elif com_kor[0].startswith('인') or com_kor[0].startswith('엔'):
        n_cnt_kor += 1
print(n_cnt_kor, s_cnt_kor)
```

```python
mbti_comment_kor = []

for us_com_kor in user_comment_kor:
    if len(us_com_kor) == 4:
        if us_com_kor[0].startswith('인') or us_com_kor[0].startswith('엔'):
            us_com_kor[0] = 'N'
            mbti_comment_kor.append(us_com_kor)
        elif us_com_kor[0].startswith('잇') or us_com_kor[0].startswith('엣'):
            us_com_kor[0] = 'S'
            mbti_comment_kor.append(us_com_kor)
```

```python
mbti_comment_kor
```

##### mbti 영어 한글 기제 합치기

```python
mbti_com_list = mbti_comment_eng + mbti_comment_kor
print(len(mbti_com_list))
mbti_com_list
```

```python
cnt_n, cnt_s = 0,0

for mbti in mbti_com_list:
    # print(mbti[0])
    if mbti[0] == 'N':
        cnt_n += 1
    elif mbti[0] == 'S':
        cnt_s += 1

cnt_n, cnt_s
```
